```python
# Método de pago y Cantidad de cuotas, obtener la suma total de las cuotas por método de pago, según por cada sucursal de las tiendas
# Calculamos las sumas por método de pago para cada tienda
Sum_Metodo_Pago1 = tienda.groupby('Método de pago')['Cantidad de cuotas'].sum().reset_index()
Sum_Metodo_Pago2 = tienda2.groupby('Método de pago')['Cantidad de cuotas'].sum().reset_index()
Sum_Metodo_Pago3 = tienda3.groupby('Método de pago')['Cantidad de cuotas'].sum().reset_index()
Sum_Metodo_Pago4 = tienda4.groupby('Método de pago')['Cantidad de cuotas'].sum().reset_index()

# Renombramos las columnas para facilitar la fusión
Sum_Metodo_Pago1.rename(columns={'Cantidad de cuotas': 'tienda'}, inplace=True)
Sum_Metodo_Pago2.rename(columns={'Cantidad de cuotas': 'tienda2'}, inplace=True)
Sum_Metodo_Pago3.rename(columns={'Cantidad de cuotas': 'tienda3'}, inplace=True)
Sum_Metodo_Pago4.rename(columns={'Cantidad de cuotas': 'tienda4'}, inplace=True)

# Fusionamos los DataFrames usando 'Método de pago' como clave
resultado = Sum_Metodo_Pago1.merge(Sum_Metodo_Pago2, on='Método de pago', how='outer')
resultado = resultado.merge(Sum_Metodo_Pago3, on='Método de pago', how='outer')
resultado = resultado.merge(Sum_Metodo_Pago4, on='Método de pago', how='outer')

# Rellenamos los valores NaN con 0 (por si algún método de pago no existe en alguna tienda)
resultado.fillna(0, inplace=True)

# Mostramos el resultado
print(resultado)
```

# 2. Ventas por categoría

```python
# En este debes calcular la cantidad de productos vendidos por categoría en cada tienda.
# La idea es agrupar los datos por categoría y contar el número de ventas de cada tipo, mostrando las categorías más populares de cada tienda.
import pandas as pd

# 1. Obtener TODOS los productos por categoría para cada tienda (incluye repetidos)
Producto_Vendidos_Categoria_Tienda1 = tienda.groupby('Categoría del Producto')['Producto'].count()
Producto_Vendidos_Categoria_Tienda2 = tienda2.groupby('Categoría del Producto')['Producto'].count()
Producto_Vendidos_Categoria_Tienda3 = tienda3.groupby('Categoría del Producto')['Producto'].count()
Producto_Vendidos_Categoria_Tienda4 = tienda4.groupby('Categoría del Producto')['Producto'].count()

# 2. Convertir cada Serie en un DataFrame con nombre correcto
df_Tienda1 = pd.DataFrame({'Tienda 1': Producto_Vendidos_Categoria_Tienda1})
df_Tienda2 = pd.DataFrame({'Tienda 2': Producto_Vendidos_Categoria_Tienda2})
df_Tienda3 = pd.DataFrame({'Tienda 3': Producto_Vendidos_Categoria_Tienda3})
df_Tienda4 = pd.DataFrame({'Tienda 4': Producto_Vendidos_Categoria_Tienda4})

# 3. Concatenar horizontalmente
Producto_Vendidos_Categoria_Union = pd.concat([df_Tienda1, df_Tienda2, df_Tienda3, df_Tienda4], axis=1)

# 4. Agregamos la columna total de suma de las tiendas
Producto_Vendidos_Categoria_Union['Total'] = Producto_Vendidos_Categoria_Union.sum(axis=1)

# 5. Ordenar por total descendente
Producto_Vendidos_Categoria_Union = Producto_Vendidos_Categoria_Union.sort_values('Total', ascending=False)

# 6. Resetear índice y mostrar
Producto_Vendidos_Categoria_Union_reset = Producto_Vendidos_Categoria_Union.reset_index()

# 7. AGREGAR AQUÍ: Crear una fila con las sumas de cada columna
suma_tienda1 = Producto_Vendidos_Categoria_Union_reset['Tienda 1'].sum()
suma_tienda2 = Producto_Vendidos_Categoria_Union_reset['Tienda 2'].sum()
suma_tienda3 = Producto_Vendidos_Categoria_Union_reset['Tienda 3'].sum()
suma_tienda4 = Producto_Vendidos_Categoria_Union_reset['Tienda 4'].sum()
suma_total = Producto_Vendidos_Categoria_Union_reset['Total'].sum()

# Crear un diccionario con los totales
fila_totales = {
    'Categoría del Producto': 'Sub Total',
    'Tienda 1': suma_tienda1,
    'Tienda 2': suma_tienda2,
    'Tienda 3': suma_tienda3,
    'Tienda 4': suma_tienda4,
    'Total': suma_total
}

# Convertir a DataFrame y concatenar
df_totales = pd.DataFrame([fila_totales])
Producto_Vendidos_Categoria_Union_completo = pd.concat([Producto_Vendidos_Categoria_Union_reset, df_totales], ignore_index=True)

# Segundo para que mi resultado salga en negrita
def negrita(texto):
    return f"\033[1m{texto}\033[0m"

print(negrita("Cantidad de productos vendidos por Categoría en cada Tienda (Ordenados por Total Descendente)"))
print("-" * 100)

# formateo para salida de miles (,)
print(Producto_Vendidos_Categoria_Union_completo.to_string(
    index=False,
    formatters={
        'Tienda 1': '{:,.0f}'.format,
        'Tienda 2': '{:,.0f}'.format,
        'Tienda 3': '{:,.0f}'.format,
        'Tienda 4': '{:,.0f}'.format,
        'Total': '{:,.0f}'.format
    }
))

print("=" * 100)
```

```python
#En este paso, debes identificar qué productos fueron los más vendidos y los menos vendidos en cada tienda. Visualiza los resultados para que quede claro qué productos destacaron en ventas en cada tienda.
import pandas as pd

# ================================================
# 1. OBTENER DATOS PARA AMBAS TABLAS
# ================================================

# Tienda 1
Productos_Vendidos_Tienda1 = tienda.groupby('Producto')['Producto'].count()
Productos_Vendidos_Tienda1 = Productos_Vendidos_Tienda1.sort_values(ascending=False)
Productos_Mas_Vendidos_Tienda1 = Productos_Vendidos_Tienda1.head(3)    # 3 MÁS vendidos
Productos_Menos_Vendidos_Tienda1 = Productos_Vendidos_Tienda1.tail(3)  # 3 MENOS vendidos

# Tienda 2
Productos_Vendidos_Tienda2 = tienda2.groupby('Producto')['Producto'].count()
Productos_Vendidos_Tienda2 = Productos_Vendidos_Tienda2.sort_values(ascending=False)
Productos_Mas_Vendidos_Tienda2 = Productos_Vendidos_Tienda2.head(3)
Productos_Menos_Vendidos_Tienda2 = Productos_Vendidos_Tienda2.tail(3)

# Tienda 3
Productos_Vendidos_Tienda3 = tienda3.groupby('Producto')['Producto'].count()
Productos_Vendidos_Tienda3 = Productos_Vendidos_Tienda3.sort_values(ascending=False)
Productos_Mas_Vendidos_Tienda3 = Productos_Vendidos_Tienda3.head(3)
Productos_Menos_Vendidos_Tienda3 = Productos_Vendidos_Tienda3.tail(3)

# Tienda 4
Productos_Vendidos_Tienda4 = tienda4.groupby('Producto')['Producto'].count()
Productos_Vendidos_Tienda4 = Productos_Vendidos_Tienda4.sort_values(ascending=False)
Productos_Mas_Vendidos_Tienda4 = Productos_Vendidos_Tienda4.head(3)
Productos_Menos_Vendidos_Tienda4 = Productos_Vendidos_Tienda4.tail(3)

# ================================================
# 2. DEFINIR FUNCIÓN PARA MOSTRAR TABLA
# ================================================

def mostrar_tabla_productos(titulo, datos_tienda1, datos_tienda2, datos_tienda3, datos_tienda4):
    """Función para mostrar tabla de productos"""
    # Definir ancho de columnas
    ancho_sucursal = 15
    ancho_producto = 30
    ancho_cantidad = 15
    ancho_total = ancho_sucursal + ancho_producto + ancho_cantidad

    # Cabeceras
    print(f'\n{titulo:^{ancho_total}}')
    print('=' * ancho_total)
    print(
        f'{"Tienda":<{ancho_sucursal}}'
        f'{"Producto":<{ancho_producto}}'
        f'{"Cantidad":>{ancho_cantidad}}'
    )
    print('-' * ancho_total)

    # Mostrar productos de cada tienda
    for producto, cantidad in datos_tienda1.items():
        print(
            f'{"Tienda 1":<{ancho_sucursal}}'
            f'{producto:<{ancho_producto}}'
            f'{cantidad:>{ancho_cantidad},}'
        )

    for producto, cantidad in datos_tienda2.items():
        print(
            f'{"Tienda 2":<{ancho_sucursal}}'
            f'{producto:<{ancho_producto}}'
            f'{cantidad:>{ancho_cantidad},}'
        )

    for producto, cantidad in datos_tienda3.items():
        print(
            f'{"Tienda 3":<{ancho_sucursal}}'
            f'{producto:<{ancho_producto}}'
            f'{cantidad:>{ancho_cantidad},}'
        )

    for producto, cantidad in datos_tienda4.items():
        print(
            f'{"Tienda 4":<{ancho_sucursal}}'
            f'{producto:<{ancho_producto}}'
            f'{cantidad:>{ancho_cantidad},}'
        )

    print('=' * ancho_total)

# ================================================
# 3. MOSTRAR AMBAS TABLAS
# ================================================

print("ANÁLISIS DE PRODUCTOS MÁS Y MENOS VENDIDOS")
print("=" * 100)

# Tabla de los 3 PRODUCTOS MÁS VENDIDOS
mostrar_tabla_productos(
    "3 PRODUCTOS MÁS VENDIDOS POR TIENDA",
    Productos_Mas_Vendidos_Tienda1,
    Productos_Mas_Vendidos_Tienda2,
    Productos_Mas_Vendidos_Tienda3,
    Productos_Mas_Vendidos_Tienda4
)

# Separador entre tablas
print("\n" + "=" * 100 + "\n")

# Tabla de los 3 PRODUCTOS MENOS VENDIDOS
mostrar_tabla_productos(
    "3 PRODUCTOS MENOS VENDIDOS POR TIENDA",
    Productos_Menos_Vendidos_Tienda1,
    Productos_Menos_Vendidos_Tienda2,
    Productos_Menos_Vendidos_Tienda3,
    Productos_Menos_Vendidos_Tienda4
)

# ================================================
# 4. INFORMACIÓN ADICIONAL (OPCIONAL)
# ================================================

print("\n" + "=" * 100)
print("RESUMEN ESTADÍSTICO")
print("-" * 100)

# Calcular estadísticas
for i, (nombre, datos) in enumerate(zip(
    ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    [Productos_Vendidos_Tienda1, Productos_Vendidos_Tienda2,
     Productos_Vendidos_Tienda3, Productos_Vendidos_Tienda4]
), 1):

    total_productos = len(datos)
    total_ventas = datos.sum()
    producto_top = datos.idxmax() if not datos.empty else "N/A"
    ventas_top = datos.max() if not datos.empty else 0
    producto_menos = datos.idxmin() if not datos.empty else "N/A"
    ventas_menos = datos.min() if not datos.empty else 0

    print(f"\n{nombre}:")
    print(f"  • Total productos diferentes: {total_productos:,}")
    print(f"  • Total unidades vendidas: {total_ventas:,}")
    print(f"  • Producto más vendido: {producto_top} ({ventas_top:,} unidades)")
    print(f"  • Producto menos vendido: {producto_menos} ({ventas_menos:,} unidades)")

print("\n" + "=" * 100)
```

# 5. Envío promedio por tienda

```python
# En este paso, debes calcular el costo de envío promedio para cada tienda. El objetivo es comprender cuánto se gasta, en promedio, en el envío de cada tienda.  Sabiendo que el cliente paga el envío.
import pandas as pd
from itertools import zip_longest

# Obtener todos los lugares únicos de todas las tiendas
todos_lugares = sorted(set(
    list(tienda['Lugar de Compra'].unique()) +
    list(tienda2['Lugar de Compra'].unique()) +
    list(tienda3['Lugar de Compra'].unique()) +
    list(tienda4['Lugar de Compra'].unique())
))

# Calcular promedio de costo de envío por lugar para cada tienda
promedios_t1 = tienda.groupby('Lugar de Compra')['Costo de envío'].mean()
promedios_t2 = tienda2.groupby('Lugar de Compra')['Costo de envío'].mean()
promedios_t3 = tienda3.groupby('Lugar de Compra')['Costo de envío'].mean()
promedios_t4 = tienda4.groupby('Lugar de Compra')['Costo de envío'].mean()

# Crear listas con los promedios para cada lugar
datos = []
for lugar in todos_lugares:
    fila = {
        'Lugares de Compra': lugar,
        'Tienda 1': promedios_t1.get(lugar, 0),
        'Tienda 2': promedios_t2.get(lugar, 0),
        'Tienda 3': promedios_t3.get(lugar, 0),
        'Tienda 4': promedios_t4.get(lugar, 0)
    }
    datos.append(fila)

# Crear DataFrame sin Promedio General
df_costos = pd.DataFrame(datos)

# Aplicar formato de miles y 2 decimales a las columnas numéricas
columnas_numericas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
for col in columnas_numericas:
    df_costos[col] = df_costos[col].apply(lambda x: f"{x:,.2f}")

# Función para texto en negrita
def negrita(texto):
    return f"\033[1m{texto}\033[0m"

# Mostrar la tabla con separadores
print(negrita("|| PROMEDIO DE COSTO DE ENVIO SEGÚN LUGAR DE COMPRA ||".center(100)))
print("-" * 100)
print(df_costos.to_string(index=False, col_space=15, justify='right'))

# Agregar salto de línea y mostrar Promedio General en negrita
print()
promedio_t1 = tienda['Costo de envío'].mean()
promedio_t2 = tienda2['Costo de envío'].mean()
promedio_t3 = tienda3['Costo de envío'].mean()
promedio_t4 = tienda4['Costo de envío'].mean()

linea_promedio = f"{'Promedio General':<20} {promedio_t1:>12,.2f} {promedio_t2:>15,.2f} {promedio_t3:>15,.2f} {promedio_t4:>15,.2f}"
print(negrita(linea_promedio))

print("=" * 100)

# Identificar máximo y mínimo
promedios = {
    'Tienda 1': promedio_t1,
    'Tienda 2': promedio_t2,
    'Tienda 3': promedio_t3,
    'Tienda 4': promedio_t4
}

tienda_max = max(promedios, key=promedios.get)
valor_max = promedios[tienda_max]

tienda_min = min(promedios, key=promedios.get)
valor_min = promedios[tienda_min]

print()
print(f"Promedio de Costo Máximo: {tienda_max} con {valor_max:,.2f}")
print(f"Promedio de Costo Mínimo: {tienda_min} con {valor_min:,.2f}")
```

```python
# Gráfico de líneas Costo de envío por tienda desplegado en los años
import pandas as pd
import matplotlib.pyplot as plt
import warnings

# Ignorar warnings
warnings.filterwarnings('ignore')

# URLs de los datos
url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

# Leer los datos con dayfirst=True para evitar el warning
TiendaGl = pd.read_csv(url)
TiendaGl2 = pd.read_csv(url2)
TiendaGl3 = pd.read_csv(url3)
TiendaGl4 = pd.read_csv(url4)

# Agregar columna de identificación de tienda
TiendaGl['Tienda'] = 'Tienda 1'
TiendaGl2['Tienda'] = 'Tienda 2'
TiendaGl3['Tienda'] = 'Tienda 3'
TiendaGl4['Tienda'] = 'Tienda 4'

# Concatenar todas las tiendas
todas_tiendas = pd.concat([TiendaGl, TiendaGl2, TiendaGl3, TiendaGl4], ignore_index=True)

# Convertir 'Fecha de Compra' a datetime con dayfirst=True
todas_tiendas['Fecha de Compra'] = pd.to_datetime(todas_tiendas['Fecha de Compra'], dayfirst=True)
todas_tiendas['Año'] = todas_tiendas['Fecha de Compra'].dt.year

# Agrupar por Tienda y Año, sumando el Costo de envío
tabla_resultado = todas_tiendas.groupby(['Tienda', 'Año'])['Costo de envío'].sum().unstack(fill_value=0)

# Ordenar columnas (años) de menor a mayor
tabla_resultado = tabla_resultado[sorted(tabla_resultado.columns)]

# Resetear el índice para que 'Tiendas' sea una columna normal
tabla_resultado_display = tabla_resultado.reset_index()

# Renombrar la columna 'Tienda' a 'Tiendas'
tabla_resultado_display = tabla_resultado_display.rename(columns={'Tienda': 'Tiendas'})

print("\n📊 TABLA DE COSTO DE ENVIO POR AÑO Y TIENDAS:".center(100))
print("-" * 80)
print(tabla_resultado_display.to_string(index=False, col_space=15, justify='center', float_format=lambda x: f'${x:,.2f}'))
print("=" * 80 + "\n")

# Crear el gráfico de líneas
plt.figure(figsize=(12, 5))

# Colores distintos para cada tienda
colores = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#FFA07A']

# Graficar cada tienda con su color
for i, tienda in enumerate(tabla_resultado.index):
    plt.plot(tabla_resultado.columns,
             tabla_resultado.loc[tienda],
             marker='o',
             linewidth=2.5,
             markersize=8,
             color=colores[i],
             label=tienda)

# Configuración del gráfico
plt.title('📈 Costo de Envío por Tienda y Año', fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Año', fontsize=12, fontweight='bold')
plt.ylabel('Costo de Envío ($)', fontsize=12, fontweight='bold')
plt.legend(title='Tiendas', fontsize=10, title_fontsize=11, loc='best')
plt.grid(True, alpha=0.3, linestyle='--')
plt.xticks(tabla_resultado.columns, fontsize=10)
plt.yticks(fontsize=10)

# Formato de números en el eje Y
ax = plt.gca()
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'${x/1e6:.1f}M'))

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from tabulate import tabulate

def negrita(texto):
    return f"\033[1m{texto}\033[0m"

# ------------------------------------------------------------------------------------------
# 1.- LOS INGRESOS TOTALES POR TIENDA
# ------------------------------------------------------------------------------------------
url_ingresos_1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url_ingresos_2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url_ingresos_3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url_ingresos_4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

# Cargar los datos para ingresos
tienda_ingresos_1 = pd.read_csv(url_ingresos_1)
tienda_ingresos_2 = pd.read_csv(url_ingresos_2)
tienda_ingresos_3 = pd.read_csv(url_ingresos_3)
tienda_ingresos_4 = pd.read_csv(url_ingresos_4)

# Ingresos totales de las tiendas
ingresos_totales_tienda_1 = tienda_ingresos_1['Precio'].sum() + tienda_ingresos_1['Costo de envío'].sum()
ingresos_totales_tienda_2 = tienda_ingresos_2['Precio'].sum() + tienda_ingresos_2['Costo de envío'].sum()
ingresos_totales_tienda_3 = tienda_ingresos_3['Precio'].sum() + tienda_ingresos_3['Costo de envío'].sum()
ingresos_totales_tienda_4 = tienda_ingresos_4['Precio'].sum() + tienda_ingresos_4['Costo de envío'].sum()

# Crear DataFrame con los ingresos
df_ingresos_totales = pd.DataFrame({
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Ingresos Totales': [ingresos_totales_tienda_1, ingresos_totales_tienda_2,
                         ingresos_totales_tienda_3, ingresos_totales_tienda_4]
})

# Formatear con comas y 2 decimales
df_ingresos_totales['Ingresos Totales'] = df_ingresos_totales['Ingresos Totales'].apply(lambda x: f"{x:,.2f}")
print(negrita("|| INGRESOS TOTALES DE LAS TIENDAS ||".center(40)))
print("-" * 40)
print(df_ingresos_totales.to_string(index=False, col_space=10, justify='left'))
print("=" * 170 + "\n")

# ------------------------------------------------------------------------------------------
# 2.- LAS CATEGORÍAS CON LAS CANTIDADES DE PRODUCTOS MÁS Y MENOS VENDIDOS
# ------------------------------------------------------------------------------------------
url_categorias_1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url_categorias_2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url_categorias_3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url_categorias_4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

# Cargar los datos para categorías
tienda_categorias_1 = pd.read_csv(url_categorias_1)
tienda_categorias_2 = pd.read_csv(url_categorias_2)
tienda_categorias_3 = pd.read_csv(url_categorias_3)
tienda_categorias_4 = pd.read_csv(url_categorias_4)

# Función para obtener las 5 categorías más y menos vendidas
def obtener_categorias_vendidas(tienda):
    productos_vendidos = tienda.groupby('Categoría del Producto')['Producto'].count()
    productos_vendidos = productos_vendidos.sort_values(ascending=False)
    mas_vendidos = productos_vendidos.head(5)  # 5 más vendidos
    menos_vendidos = productos_vendidos.tail(5)  # 5 menos vendidos
    return mas_vendidos, menos_vendidos

# Obtener los datos para cada tienda
mas_vendidos_categoria_1, menos_vendidos_categoria_1 = obtener_categorias_vendidas(tienda_categorias_1)
mas_vendidos_categoria_2, menos_vendidos_categoria_2 = obtener_categorias_vendidas(tienda_categorias_2)
mas_vendidos_categoria_3, menos_vendidos_categoria_3 = obtener_categorias_vendidas(tienda_categorias_3)
mas_vendidos_categoria_4, menos_vendidos_categoria_4 = obtener_categorias_vendidas(tienda_categorias_4)

# Crear DataFrames para más vendidos
df_mas_vendidos_categorias = pd.DataFrame({
    "Categoría del Producto": mas_vendidos_categoria_1.index,
    "Tienda 1": mas_vendidos_categoria_1.values,
    "Tienda 2": mas_vendidos_categoria_2.reindex(mas_vendidos_categoria_1.index).fillna(0).astype(int),
    "Tienda 3": mas_vendidos_categoria_3.reindex(mas_vendidos_categoria_1.index).fillna(0).astype(int),
    "Tienda 4": mas_vendidos_categoria_4.reindex(mas_vendidos_categoria_1.index).fillna(0).astype(int)
})

# Crear DataFrames para menos vendidos
df_menos_vendidos_categorias = pd.DataFrame({
    "Categoría del Producto": menos_vendidos_categoria_1.index,
    "Tienda 1": menos_vendidos_categoria_1.values,
    "Tienda 2": menos_vendidos_categoria_2.reindex(menos_vendidos_categoria_1.index).fillna(0).astype(int),
    "Tienda 3": menos_vendidos_categoria_3.reindex(menos_vendidos_categoria_1.index).fillna(0).astype(int),
    "Tienda 4": menos_vendidos_categoria_4.reindex(menos_vendidos_categoria_1.index).fillna(0).astype(int)
})

# Calcular el total para más vendidos
df_mas_vendidos_categorias["Total"] = df_mas_vendidos_categorias[["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]].sum(axis=1)

# Calcular el total para menos vendidos
df_menos_vendidos_categorias["Total"] = df_menos_vendidos_categorias[["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]].sum(axis=1)

# Imprimir los resultados en una sola tabla
print('\n')
print(negrita("|| CATEGORÍAS DE PRODUCTOS MÁS VENDIDOS" + " " * 20 + "CATEGORÍAS DE PRODUCTOS MENOS VENDIDOS ||".center(100)))
print("-" * 170)

# Convertir los DataFrames a tablas con tabulate
tabla_mas_vendidos_categorias = tabulate(df_mas_vendidos_categorias, headers='keys', tablefmt='pretty', showindex=False)
tabla_menos_vendidos_categorias = tabulate(df_menos_vendidos_categorias, headers='keys', tablefmt='pretty', showindex=False)

# Dividir las líneas de las tablas
lineas_mas_vendidos_categorias = tabla_mas_vendidos_categorias.split('\n')
lineas_menos_vendidos_categorias = tabla_menos_vendidos_categorias.split('\n')

# Imprimir las tablas lado a lado
for linea_mas, linea_menos in zip(lineas_mas_vendidos_categorias, lineas_menos_vendidos_categorias):
    print(linea_mas + " " * 10 + linea_menos)

print("=" * 170)

# ------------------------------------------------------------------------------------------
# 3.- LAS CALIFICACIONES PROMEDIO DE LOS CLIENTES POR TIENDA
# ------------------------------------------------------------------------------------------
url_calificaciones_1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url_calificaciones_2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url_calificaciones_3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url_calificaciones_4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

# Cargar los datos para calificaciones
tienda_calificaciones_1 = pd.read_csv(url_calificaciones_1)
tienda_calificaciones_2 = pd.read_csv(url_calificaciones_2)
tienda_calificaciones_3 = pd.read_csv(url_calificaciones_3)
tienda_calificaciones_4 = pd.read_csv(url_calificaciones_4)

# Contar calificaciones para cada sucursal (1 al 4)
conteo_calificaciones_1 = tienda_calificaciones_1['Calificación'].value_counts().reindex([1, 2, 3, 4], fill_value=0)
conteo_calificaciones_2 = tienda_calificaciones_2['Calificación'].value_counts().reindex([1, 2, 3, 4], fill_value=0)
conteo_calificaciones_3 = tienda_calificaciones_3['Calificación'].value_counts().reindex([1, 2, 3, 4], fill_value=0)
conteo_calificaciones_4 = tienda_calificaciones_4['Calificación'].value_counts().reindex([1, 2, 3, 4], fill_value=0)

# Crear un DataFrame con las calificaciones como filas y las tiendas como columnas
df_calificaciones = pd.DataFrame({
    'Tienda 1': conteo_calificaciones_1,
    'Tienda 2': conteo_calificaciones_2,
    'Tienda 3': conteo_calificaciones_3,
    'Tienda 4': conteo_calificaciones_4
})

# Calcular el total por calificación
df_calificaciones['Total'] = df_calificaciones.sum(axis=1)

# Imprimir la tabla con formato
print('\n')
print(negrita("|| LAS CALIFICACIONES DE LOS CLIENTES POR TIENDA ||".center(40)))
print('-' * 70)
print(f'{"Calificación":<12}' + f'{"Tienda 1":>12}' + f'{"Tienda 2":>12}' + f'{"Tienda 3":>12}' + f'{"Tienda 4":>12}' + f'{"Total":>12}')
print('-' * 70)
for calificacion in [1, 2, 3, 4]:
    print(
        f'{calificacion:<12}' +
        f'{df_calificaciones.loc[calificacion, "Tienda 1"]:>12,}' +
        f'{df_calificaciones.loc[calificacion, "Tienda 2"]:>12,}' +
        f'{df_calificaciones.loc[calificacion, "Tienda 3"]:>12,}' +
        f'{df_calificaciones.loc[calificacion, "Tienda 4"]:>12,}' +
        f'{df_calificaciones.loc[calificacion, "Total"]:>12,}'
    )
print('=' * 170)

# ------------------------------------------------------------------------------------------
# 4.- LOS PRODUCTOS MÁS Y MENOS VENDIDOS POR TIENDA
# ------------------------------------------------------------------------------------------
url_productos_1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url_productos_2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url_productos_3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url_productos_4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

# Cargar los datos para productos
tienda_productos_1 = pd.read_csv(url_productos_1)
tienda_productos_2 = pd.read_csv(url_productos_2)
tienda_productos_3 = pd.read_csv(url_productos_3)
tienda_productos_4 = pd.read_csv(url_productos_4)

# Función para obtener los 5 productos más y menos vendidos por tienda
def obtener_productos_vendidos(tienda):
    productos_vendidos = tienda['Producto'].value_counts()
    mas_vendidos = productos_vendidos.head(5)  # 5 productos más vendidos
    menos_vendidos = productos_vendidos.tail(5)  # 5 productos menos vendidos
    return mas_vendidos, menos_vendidos

# Obtener los datos para cada tienda
mas_vendidos_productos_1, menos_vendidos_productos_1 = obtener_productos_vendidos(tienda_productos_1)
mas_vendidos_productos_2, menos_vendidos_productos_2 = obtener_productos_vendidos(tienda_productos_2)
mas_vendidos_productos_3, menos_vendidos_productos_3 = obtener_productos_vendidos(tienda_productos_3)
mas_vendidos_productos_4, menos_vendidos_productos_4 = obtener_productos_vendidos(tienda_productos_4)

# Crear DataFrames para cada tienda (más vendidos)
df_mas_vendidos_productos = {
    "Tienda 1": pd.DataFrame({"Producto": mas_vendidos_productos_1.index, "Tienda 1": mas_vendidos_productos_1.values}),
    "Tienda 2": pd.DataFrame({"Producto": mas_vendidos_productos_2.index, "Tienda 2": mas_vendidos_productos_2.values}),
    "Tienda 3": pd.DataFrame({"Producto": mas_vendidos_productos_3.index, "Tienda 3": mas_vendidos_productos_3.values}),
    "Tienda 4": pd.DataFrame({"Producto": mas_vendidos_productos_4.index, "Tienda 4": mas_vendidos_productos_4.values}),
}

# Crear DataFrames para cada tienda (menos vendidos)
df_menos_vendidos_productos = {
    "Tienda 1": pd.DataFrame({"Producto": menos_vendidos_productos_1.index, "Tienda 1": menos_vendidos_productos_1.values}),
    "Tienda 2": pd.DataFrame({"Producto": menos_vendidos_productos_2.index, "Tienda 2": menos_vendidos_productos_2.values}),
    "Tienda 3": pd.DataFrame({"Producto": menos_vendidos_productos_3.index, "Tienda 3": menos_vendidos_productos_3.values}),
    "Tienda 4": pd.DataFrame({"Producto": menos_vendidos_productos_4.index, "Tienda 4": menos_vendidos_productos_4.values}),
}

# Convertir cada DataFrame a una tabla con tabulate
tablas_mas_vendidos_productos = {
    tienda: tabulate(df, headers='keys', tablefmt='pretty', showindex=False)
    for tienda, df in df_mas_vendidos_productos.items()
}

tablas_menos_vendidos_productos = {
    tienda: tabulate(df, headers='keys', tablefmt='pretty', showindex=False)
    for tienda, df in df_menos_vendidos_productos.items()
}

# Dividir las tablas en líneas para alinearlas
lineas_mas_vendidos_productos = [tabla.split('\n') for tabla in tablas_mas_vendidos_productos.values()]
lineas_menos_vendidos_productos = [tabla.split('\n') for tabla in tablas_menos_vendidos_productos.values()]

# Imprimir los productos más vendidos en una sola fila
print('\n')
print(negrita("|| PRODUCTOS MÁS VENDIDOS POR TIENDA ||".center(150)))
print("-" * 150)

# Encontrar el máximo número de líneas en las tablas de más vendidos
max_lineas_mas = max(len(lineas) for lineas in lineas_mas_vendidos_productos)

for i in range(max_lineas_mas):
    linea_completa = ""
    for j, tienda in enumerate(df_mas_vendidos_productos.keys()):
        linea = lineas_mas_vendidos_productos[j][i] if i < len(lineas_mas_vendidos_productos[j]) else " " * 30
        linea_completa += linea + " " * 7  # Separación de 7 espacios entre tablas
    print(linea_completa)

# Imprimir los productos menos vendidos en una sola fila
print("\n" + "-" * 150 + "\n")
print(negrita("|| PRODUCTOS MENOS VENDIDOS POR TIENDA ||".center(150)))
print("-" * 150)

# Encontrar el máximo número de líneas en las tablas de menos vendidos
max_lineas_menos = max(len(lineas) for lineas in lineas_menos_vendidos_productos)

for i in range(max_lineas_menos):
    linea_completa = ""
    for j, tienda in enumerate(df_menos_vendidos_productos.keys()):
        linea = lineas_menos_vendidos_productos[j][i] if i < len(lineas_menos_vendidos_productos[j]) else " " * 30
        linea_completa += linea + " " * 7  # Separación de 7 espacios entre tablas
    print(linea_completa)

print("\n" + "=" * 150)

# ------------------------------------------------------------------------------------------
# 5.- EL COSTO DE ENVIO PROMEDIO POR CADA TIENDA
# ------------------------------------------------------------------------------------------
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import warnings

# Ignorar warnings
warnings.filterwarnings('ignore')

# URLs de los datos
url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

# Leer los datos con dayfirst=True para evitar el warning
TiendaGl = pd.read_csv(url)
TiendaGl2 = pd.read_csv(url2)
TiendaGl3 = pd.read_csv(url3)
TiendaGl4 = pd.read_csv(url4)

# Agregar columna de identificación de tienda
TiendaGl['Tienda'] = 'Tienda 1'
TiendaGl2['Tienda'] = 'Tienda 2'
TiendaGl3['Tienda'] = 'Tienda 3'
TiendaGl4['Tienda'] = 'Tienda 4'

# Concatenar todas las tiendas
todas_tiendas = pd.concat([TiendaGl, TiendaGl2, TiendaGl3, TiendaGl4], ignore_index=True)

# Convertir 'Fecha de Compra' a datetime con dayfirst=True
todas_tiendas['Fecha de Compra'] = pd.to_datetime(todas_tiendas['Fecha de Compra'], dayfirst=True)
todas_tiendas['Año'] = todas_tiendas['Fecha de Compra'].dt.year

# Agrupar por Tienda y Año, sumando el Costo de envío
tabla_resultado = todas_tiendas.groupby(['Tienda', 'Año'])['Costo de envío'].sum().unstack(fill_value=0)

# Ordenar columnas (años) de menor a mayor
tabla_resultado = tabla_resultado[sorted(tabla_resultado.columns)]

# Resetear el índice para que 'Tiendas' sea una columna normal
tabla_resultado_display = tabla_resultado.reset_index()

# Renombrar la columna 'Tienda' a 'Tiendas'
tabla_resultado_display = tabla_resultado_display.rename(columns={'Tienda': 'Tiendas'})

print("\n📊 TABLA DE COSTO DE ENVIO POR AÑO Y TIENDAS:".center(100))
print("-" * 80)
print(tabla_resultado_display.to_string(index=False, col_space=15, justify='center', float_format=lambda x: f'${x:,.2f}'))
print("=" * 80 + "\n")
```
