```python
# Borrar el 1er registro de empleados
from sqlalchemy import text

# Borrar el 1er registro de empleados
Query = 'DELETE FROM empleados WHERE ID_Cliente = 5008804'

with engine.connect() as conexion:
    resultado = conexion.execute(text(Query))
    conexion.commit()  # Confirmar los cambios

    # Verificar si se eliminó algún registro
    if resultado.rowcount > 0:
        print(f"Registro eliminado exitosamente! ({resultado.rowcount} fila(s) afectada(s))")
    else:
        print("No se encuentra el valor a eliminar")
```

```python
# Actualizando registros

Query = 'UPDATE FROM empleados SET Categoria_de_renta = "Independiente" WHERE ID_Cliente = 5008804'
try:
  r_set=engine.execute(Query)
  engine.commit()
except
  print("No se pudo actualizar")
  engine.rollback()
  engine.close()
 else
  print("Actualizado exitosamente")
```

```python
# trayendo a leer la data de clientes_banco pero como datos sql
import pandas as pd
from sqlalchemy import create_engine, MetaData, Table, inspect , text

# Leer el CSV
cliente_banco_df = pd.read_csv('/content/clientes_banco.csv')

# Crear una base de datos SQLite en memoria (o en archivo)
engine = create_engine('sqlite:///BancoDB.db')

# Cargar el DataFrame a una tabla SQL
cliente_banco_df.to_sql('Clientes', engine, if_exists='replace', index=False)

print("CSV cargado exitosamente a SQLAlchemy")
```

```python
# Crear un nuevo registro de cliente siguiendo las especificaciones a continuación:
# ID_cliente: 6850985
# Edad: 33
# Grado_estudio: Doctorado
# Estado_civil: Soltero
# Tamaño_familia: 1
# Categoria_de_renta: Empleado
# Ocupacion: TI
# Años_empleado: 2
# Rendimiento_anual: 290000
# Tiene_carro: 0
# Vivienda: Casa/Departamento propio

query = '''
INSERT INTO Clientes (ID_cliente, Edad, Grado_estudio, Estado_civil, Tamaño_familia, Categoria_de_renta, Ocupacion, Años_empleado, Rendimiento_anual, Tiene_carro, Vivienda)
VALUES (6850985, 33, "Doctorado", "Soltero", 1, "Empleado", "TI", 2, 290000,0, "Casa/Departamento propio")
'''
with engine.connect() as conexion:
    resultado = conexion.execute(text(query))
    conexion.commit()
    # Verificar que se ingreso correctamente
    if resultado.rowcount > 0:
      print(f"Registro ingresado exitosamente! ({resultado.rowcount})")
    else:
      print("No se pudo ingresar el registro")
```
