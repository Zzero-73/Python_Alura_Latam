# **Para Sheeets Google**

## Opción 1: Usando la URL directamente (más fácil)

```python
import pandas as pd

# URL del Google Sheet (debe estar en modo "Cualquiera con el enlace")
url = 'https://docs.google.com/spreadsheets/d/TU_ID_DEL_SHEET/edit#gid=0'

# Convertir a formato CSV exportable
url_csv = url.replace('/edit#gid=', '/export?format=csv&gid=')

# Leer
df = pd.read_excel(url_csv)
```

## Opción 2: Formato más simple

```python
# Si tienes la URL completa
sheet_id = 'TU_ID_DEL_SHEET'  # La parte después de /d/ en la URL
gid = '0'  # ID de la hoja específica (después de gid=)

url = f'https://docs.google.com/spreadsheets/d/{sheet_id}/export?format=csv&gid={gid}'
df = pd.read_csv(url)
```
