```python
# Leer JSON subido
import pandas as pd
ruta = '/content/drive/MyDrive/Colab Notebooks/datos_hosting.json'
df = pd.read_json(ruta)
df
```

```python
# normalizando el json
Datos = pd.json_normalize(df['info_inmuebles'])
Datos.head(5)
```

## **Datos Numéricos**

```python
Columnas = list(Datos.columns)
Columnas
```

```python
# Separando los datos lista dentro del diccionado para Datos
Datos = Datos.explode(Columnas[3:])
Datos
```

```python
# Resetear el indice
Datos.reset_index(drop=True, inplace=True)
Datos.head(5)
```
