```python
# Calcular el promedio de habitaciones por departamento.
import pandas as pd
AlquilerURL = 'https://gist.githubusercontent.com/ahcamachod/a572cfcc2527046db93101f88011b26e/raw/ffb13f45a79d31223e645611a119397dd127ee3c/alquiler.csv'
Alquiler = pd.read_csv(AlquilerURL, sep = ';')
Promedio = Alquiler.query("Tipo == 'Departamento'")['Habitaciones'].mean()
print(f"Promedio de habitaciones en departamentos: {Promedio:.1f}")
Total_Departamentos = Alquiler.query("Tipo == 'Departamento'")['Habitaciones'].count()
print(f"Total de departamentos analizados: {Total_Departamentos}")
```

```python
# Verificar cuántas colonias únicas existen en nuestra base de datos.
Colonias_Unicas = Alquiler.Colonia.unique()
#print(*Colonias_Unicas, sep='\n')  # El * desempaqueta la list
for i,Colonia in enumerate(Colonias_Unicas,1):
    print(f"{i} {Colonia} ")
#
```

```python
# Analizar qué colonias tienen el promedio de alquiler más alto.
Alquiler.groupby('Colonia').Valor.mean().sort_values(ascending = False).map(lambda x: f'{x:,.2f}')
```

## **fillna() - Rellena valores faltantes**
Alquiler.fillna(0) reemplaza todos los valores NaN/None en el DataFrame con el valor que le especifiques (en este caso, 0).

```python
Alquiler = Alquiler.fillna(0)
```

```python
#Remover departamentos con valor igual a 0
Alquiler.query('Valor ==0 | Condominio ==0')
```

```python
# Trayendo solo los indices de la lista con valor 0 y condominio 0
Alquiler.query('Valor ==0 | Condominio ==0').index
```

```python
# Removemos las filas de la lista con valor 0 y condominio 0
Alquiler = Alquiler.drop(Alquiler.query('Valor ==0 | Condominio ==0').index)
```
